```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('data/ratings.dat', sep='::', header=None, 
                 names=['user_id', 'movie_id', 'rating', 'timestamp'], 
                 encoding='iso-8859-1', engine='python')
```

```python
train, test = train_test_split(df, test_size=0.2)
train.to_csv('data/ml1m_train.csv', index=False)
test.to_csv('data/ml1m_test.csv', index=False)
```

```python
print("Total", df.shape[0], len(df.user_id.unique()), len(df.movie_id.unique()))
print("Train", train.shape[0], len(train.user_id.unique()), len(train.movie_id.unique()))
print("Test ", test.shape[0], len(test.user_id.unique()), len(test.movie_id.unique()))
```
